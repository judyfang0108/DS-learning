### TODO #3: 利用 Seaborn 畫箱型圖，並進行說明
- 觀察每一個 features 的數值分佈
- 可以用來決定是否要對資料的數值進行標準化

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plot a boxplot with Seaborn
fig,ax=plt.subplots(2,7,figsize=(20,10))
ax = ax.flatten()#把[[0,0],[0,1],[1,0],[1,1],[2,0]...的2x7二維ndarray降成0,1,2,3,...的1x14一維ndarray)
index=0
for col in df.columns:
    sns.boxplot(y=df[col], ax=ax[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

從圖中可以發現CRIM,ZN,RM,B都存在較明顯的outliers

```python
#顯示每一個feature存在outlier的程度
#k:keys,v:values
for k, v in df.items():
        q1 = v.quantile(0.25)
        q3 = v.quantile(0.75)
        irq = q3 - q1
        v_col = v[(v <= q1 - 1.5 * irq) | (v >= q3 + 1.5 * irq)]
        perc = np.shape(v_col)[0] * 100.0 / np.shape(df)[0]
        print("Column %s outliers = %.2f%%" % (k, perc))

#移除MEDV的outlier
df=df[~(df['MEDV']>=50)]
```

```python
# Plot a distplot with Seaborn to see the disribution
fig,ax=plt.subplots(2,7,figsize=(20,10))
ax = ax.flatten()#把[[0,0],[0,1],[1,0],[1,1],[2,0]...的2x7二維ndarray降成0,1,2,3,...的1x14一維ndarray)
index=0
for col in df.columns:
    sns.distplot(df[col], ax=ax[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```



*   CRIM,ZN,B都是highly skew distributions

## 訓練模型前的準備
### TODO #7: 切分資料
- 利用 train_test_split，將資料切成 train set 和 test set
- 設定 train set 的比例為全部資料的 80%

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test =train_test_split(X,Y,train_size=0.8,random_state=42)
print(X_train.shape) 
print(X_test.shape) 
print(Y_train.shape) 
print(Y_test.shape) 
```

## 訓練模型並進行評估

### TODO #8: 訓練 Linear Regression 模型，並說明訓練結果
- 利用 scikit-learn，將資料 fit 在 LinearRegression
- 將結果繪製成散佈圖 (scatter plot)，其中 X軸 為 Y_pred、Y軸為 Y_true

```python
# Modeling
from sklearn import metrics
from sklearn.linear_model import LinearRegression

model_LR=LinearRegression()
model_LR.fit(X_train,Y_train)
Y_pred=model_LR.predict(X_test)
print('R2:',metrics.r2_score(Y_test, Y_pred))
print('MSE:',metrics.mean_squared_error(Y_test, Y_pred))
```

### TODO #9: 訓練 Random Forest Regressor 模型

```python
from sklearn.ensemble import RandomForestRegressor
# Modeling
model_RFR=RandomForestRegressor(random_state=1012)
model_RFR.fit(X_train,Y_train)
Y_pred=model_RFR.predict(X_test)
print('R2:',metrics.r2_score(Y_test, Y_pred))
print('MSE:',metrics.mean_squared_error(Y_test, Y_pred))
```

```python
# plotting the Y_test vs Y_pred
sns.scatterplot(Y_pred,Y_test)
```

### TODO #10: 參數調整，並進行說明
- 請嘗試使用 Grid search 來進行調參
    - 可以自己寫 for-loop
    - 或是使用 sklearn.model_selection 中的 GridSearchCV 函式

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor

#定義參數調整知函式
def find_best_model(X_train,Y_train,X_test):
  #選用LinearRegression,RandomForestRegressor,SVR,GradientBoostingRegressor作model
  #並找出每個model的best esimator
  models={
    'LinearRegression':{
        'model':LinearRegression(),
        'params_grid':{}
    },
    'RandomForestRegressor':{
        'model':RandomForestRegressor(random_state=1012),
        'params_grid':[{'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},{'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4],'random_state':[1012]}]
    },
    'SVR':{
        'model':SVR(),
        'params_grid':[{"C": [1e0, 1e1, 1e2, 1e3], "gamma": np.logspace(-2, 2, 5)}]
    },
    'GradientBoostingRegressor':{
        'model':GradientBoostingRegressor(),
        'params_grid':[{'n_estimators':[100], 'learning_rate': [0.1], 'max_depth':[6],'min_samples_leaf':[3], 'max_features':[3],'random_state':[1012]}]
    }
  }
  scores={}#存放每一個經過參數調整model的score
  errors={}#存放每一個經過參數調整model的error
  Y_predict={}#存放每一個經過參數調整model所得的Y_pred

  #得到最好的estimator用，拿來訓練模型
  for model_names,model_params in models.items():
    grid_search=GridSearchCV(model_params['model'], param_grid=model_params['params_grid'], cv=5,scoring='r2', n_jobs=2)
    grid_search.fit(X_train,Y_train)
    best_model=grid_search.best_estimator_
    best_model.fit(X_train,Y_train)
    Y_pred=best_model.predict(X_test)

    #將結果保存
    scores[model_names]=metrics.r2_score(Y_test, Y_pred)
    errors[model_names]=metrics.mean_squared_error(Y_test, Y_pred)
    Y_predict[model_names]=Y_pred
  return scores, errors,Y_predict

scores_map1=[]
errors_map1=[]
predict1=[]
scores_map1,errors_map1,predict1=find_best_model(X_train,Y_train,X_test)
Y_pred=pd.Series(predict1)
scores_select=pd.Series(scores_map1)
scores_select
```

```python
#linear regression model best estimator
print('R2:',scores_map1['LinearRegression'])
print('MSE:',errors_map1['LinearRegression'])
```

```python
#RandomForestRegressor model best estimator
print('R2:',scores_map1['RandomForestRegressor'])
print('MSE:',errors_map1['RandomForestRegressor'])
```
